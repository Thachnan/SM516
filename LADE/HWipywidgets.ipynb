### การสร้างวิดเจ็ต

ตอนนี้เราได้ติดตั้งแพ็คเกจแล้วเราสามารถสร้างอินสแตนซ์ของเครื่องมือใด ๆ ในสมุดบันทึกของเราได้ทันที คุณสามารถคลิกที่นี่เพื่อตรวจสอบรายการของวิดเจ็ตที่มีอยู่ในปัจจุบันพร้อมรหัสเพื่อเริ่มต้นพวกเขา ลองดูตัวอย่างที่กำหนดวิดเจ็ตบางอย่าง:

#### ฟังก์ชั่นที่ใช้งานสะดวกเพื่อใช้กับวิดเจ็ต
จาก IPython.display แสดงการนำเข้า Markdown, clear_output
#### แพ็คเกจวิดเจ็ต
อิมพอร์ต ipywidgets เป็นวิดเจ็ต

#### กำหนดวิดเจ็ตบางอย่าง

```python
text = widgets.Text (                       
       value = 'My Text'
       description = 'ชื่อ',)
calendar = widgets.DatePicker (
           description = 'เลือกวันที่')
slider = widgets.FloatSlider (
         ค่า = 1
         นาที = 0
         สูงสุด = 10.0
         ขั้นตอน = 0.1)
menu = widgets.Dropdown (
       ตัวเลือก = ['red', 'blue', 'green'],
       value = 'สีแดง'
       คำอธิบาย = 'สี:')
checkbox = widgets.Checkbox (
           description = 'Check to invert',)
```

ดังที่แสดงไว้ด้านบน (และในกรณีส่วนใหญ่) การเริ่มต้นวิดเจ็ตเกี่ยวข้องกับการเรียกใช้ฟังก์ชั่นที่เรียบง่ายและระบุข้อโต้แย้งที่อธิบายตนเองบางอย่างมันตรงไปตรงมามาก ที่นี่เรากำหนดเมนูแบบเลื่อนลงตัวเลื่อนช่องทำเครื่องหมายหน้าต่างข้อความและปฏิทินในสมุดบันทึกของเรา คุณสามารถแสดงใด ๆ ของพวกเขาโดยการเรียกวัตถุบนเซลล์:

*** เมนู ***

แต่เพื่อที่จะแสดงมันทั้งหมดเราจำเป็นต้องใช้วิดเจ็ตคอนเทนเนอร์

#### สร้างตู้คอนเทนเนอร์

คนที่พบมากที่สุดจะถูกกำหนดโดยใช้ฟังก์ชั่น VBox และ HBox และให้บริการเพื่อแสดงเครื่องมือของเราในแนวตั้งหรือแนวนอนตามลำดับ พวกเขาคาดหวังอาร์กิวเมนต์เดียว: รายการของวิดเจ็ต

```python
box = widgets.VBox ([ข้อความ, ตัวเลื่อน, เมนู, ปฏิทิน, ช่องทำเครื่องหมาย])
```

*** กล่อง ***

เนื่องจากคอนเทนเนอร์ของ VBox และ HBox นั้นเป็นวิดเจ็ตเราจึงสามารถใช้มันภายในกล่องเดียวกัน ลองดูตัวอย่างจำลอง: <br>

*** วิดเจ็ต. HBox ([กล่อง, กล่อง]) ***

หรือเราสามารถใช้คอนเทนเนอร์ประเภทอื่นที่อยู่ด้านบน มาดูคอนเทนเนอร์ของแท็บ ในขณะที่ซับซ้อนขึ้นเล็กน้อยมันช่วยให้เราใช้พื้นที่ได้ดีขึ้น:

```python
# กำหนดรายการที่มีเนื้อหาของหน้าต่างของเรา
children = [box, box]
```

```python
# การเริ่มต้นแท็บ
tab = widgets.Tab ()
# การตั้งค่าหน้าต่างแท็บ
tab.children = children
```

```python
# การเริ่มต้นแท็บ
tab = widgets.Tab ()
# การตั้งค่าหน้าต่างแท็บ
tab.children = children
```

### วิดเจ็ตถือคุณสมบัติ

ตอนนี้เรารู้วิธีสร้างอินสแตนซ์และจัดวางวิดเจ็ตของเรามันสำคัญมากที่ต้องคำนึงว่าแต่ละวิดเจ็ตเหล่านี้ถูกกำหนดเป็นวัตถุและมีคุณสมบัติของตัวเอง คุณสามารถพูดได้ว่าแต่ละวิดเจ็ตมีชุดของตัวเองที่แตกต่างกันตัวอย่างเช่นเมนูมีคุณสมบัติที่เรียกว่าตัวเลือก:

*** menu.options ***

การกำหนดแอตทริบิวต์นี้ใหม่จะเปลี่ยนตัวเลือกปัจจุบันของวิดเจ็ตของเรา <br>

menu.options = ('แดง', 'น้ำเงิน', 'เขียว', 'ดำ') <br>

*** เมนู ***

ด้วยข้อยกเว้นของคอนเทนเนอร์แอตทริบิวต์ที่ใช้ร่วมกันระหว่างวิดเจ็ตทั้งหมดที่เรากำหนดไว้ด้านบนคือค่าและอาจเป็นประโยชน์ที่สุด มันหมายถึงสิ่งที่ควร: เรียกมันว่ามันคืนค่าปัจจุบันของวิดเจ็ต

```python
# มูลค่าปัจจุบันของข้อความช่องทำเครื่องหมายและปฏิทิน
พิมพ์ ('ข้อความวิดเจ็ตมีค่า {}, จาก {}'. รูปแบบ (text.value, ประเภท (text.value))
พิมพ์ ('ช่องทำเครื่องหมายวิดเจ็ตมีค่า {}, จาก {}'. รูปแบบ (checkbox.value, ประเภท (checkbox.value))
พิมพ์ ('Widget ปฏิทินมีค่า {}, จาก {}'. รูปแบบ (calendar.value, ประเภท (calendar.value))
```

ความสามารถในการเรียกคืนคุณค่าของวิดเจ็ตของเราคือสิ่งที่ช่วยให้คุณสามารถโต้ตอบกับโค้ดของคุณได้ ลองดูตัวอย่างจำลองโดยใช้เมนู

```python
# เราจะใช้ค่าของเมนูเพื่อแสดงพล็อตและชื่อของมัน
นำเข้า matplotlib.pyplot เป็น plt
plt.plot ([1,2,3], color = menu.value)
plt.title ('A {} line'.format (menu.value))
plt.show ()
```

การเปลี่ยนค่าของเมนูวิดเจ็ตของเราและการเรียกใช้โค้ดข้างต้นอีกครั้งจะเปลี่ยนชื่อและสีของเส้น เราสามารถทำได้เพราะตัวเลือกของเมนูตรงกับค่าที่ถูกต้องสำหรับสีอาร์กิวเมนต์ของ plt.plot

สิ่งนี้เน้นถึงคุณสมบัติที่มีประโยชน์ของวิดเจ็ต: คุณสามารถใช้มันเพื่อควบคุมสิ่งต่างๆโดยไม่ต้องพิมพ์ออกมาอย่างชัดเจน

### การสร้างปุ่ม

คุณอาจสังเกตเห็นว่าสิ่งที่เราทำด้านบนเกี่ยวข้องกับการคลิกที่วิดเจ็ตและเลือกสิ่งต่างๆ แต่เพื่อให้สามารถกระตุ้นสิ่งที่เรากำลังเรียกใช้งานโค้ด เราต้องการวิดเจ็ตที่สามารถกระตุ้นการตอบสนองและรันโค้ดเราต้องการพลังของปุ่ม <br>

การสร้างอินสแตนซ์ของปุ่มนั้นคล้ายกับวิธีที่เรากำหนดวิดเจ็ต:

```python
button = widgets.Button (คำอธิบาย = 'ปุ่มของฉัน')
```

แต่เพื่อให้ปุ่มของเราใช้งานได้เราต้องกำหนดว่าจะเกิดอะไรขึ้นเมื่อเรากด กล่าวง่ายๆก็คือ:

* การกำหนดวิดเจ็ตเอาต์พุตที่แสดงการตอบกลับ
* การกำหนดฟังก์ชั่นที่จะดำเนินการ
* การเชื่อมโยงปุ่มของเราเอาต์พุตและฟังก์ชั่นเข้าด้วยกัน

นี่เป็นตัวอย่างเล็กน้อยที่ฉันใช้ตลอดเวลาเมื่อสร้างปุ่มใหม่: นี่คือเทมเพลตสำหรับตัวอย่างของเรา

```python
button = widgets.Button (คำอธิบาย = 'ปุ่มของฉัน')
out = widgets.Output ()
def on_button_clicked (_):
      # "ฟังก์ชันเชื่อมโยงกับเอาต์พุต"
      โดยไม่ต้องออก:
          # จะเกิดอะไรขึ้นเมื่อเรากดปุ่ม
          clear_output ()
          พิมพ์ ('มีบางอย่างเกิดขึ้น!')
# การเชื่อมโยงปุ่มและฟังก์ชั่นร่วมกันโดยใช้วิธีการของปุ่ม
button.on_click (on_button_clicked)
# แสดงปุ่มและเอาท์พุทมันเข้าด้วยกัน
widgets.VBox ([ปุ่มออก])
```

ฟังก์ชั่น clear_output รีเฟรชหน้าจอทุกครั้งที่กดปุ่ม หากเราไม่ใช้ฟังก์ชั่นนี้การกดปุ่มหลาย ๆ ครั้งจะพิมพ์ ething เกิดอะไรขึ้น! ’หลายครั้งซึ่งดูเหมือนจะไม่เป็นที่ต้องการ วิดเจ็ต Output เป็นสิ่งที่น่าสนใจจริงๆ: เราสามารถใช้มันเพื่อแสดงข้อความสั่งพิมพ์ (อย่างที่เราทำด้านบน), แปลง, วิดีโอ, รูปภาพและโดยทั่วไป

สิ่งใดก็ตามที่แสดงได้ดีในสมุดบันทึก Jupyter จะแสดงได้ดีในวิดเจ็ตเอาต์พุต

ลองดูตัวอย่างเล็ก ๆ น้อย ๆ เกี่ยวกับวิธีที่เราสามารถใช้มันเพื่อแสดง Markdown

จาก IPython.display นำเข้า Markdown

```python
# การใช้เอาต์พุตเพื่อแสดง Markdown
markdown_out = widgets.Output ()
ด้วย markdown_out:
    display (Markdown ('Slider value คือ $ {} $'. format (slider.value)))
```

*** markdown_out ***
